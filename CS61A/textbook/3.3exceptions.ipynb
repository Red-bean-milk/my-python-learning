```python
try:
    x = 1/0
except ZeroDivisionError as e:   # 异常是一个对象实例，ZeroDivisionError是一个类
    print('handling a',type(e))
    x = 0    
```

```python
def invert(x):
    result = 1/x
    print('Never printed if x is 0')
    return result

def invert_safe(x):
    try:
        return invert(x)
    except ZeroDivisionError as e:
        return str(e)

invert_safe(0)
```

```python
class IterImproveError(Exception):
    def __init__(self,last_guess):
        self.last_guess = last_guess
```

```python
def improve(update,done,guess=1,max_updates=1000):
    k = 0 # 迭代次数
    try:
        while not done(guess) and k < max_updates:
            guess = update(guess)
            k = k + 1
        return guess
    except ValueError:    # 给sqrt传入的参数为负数时会产生ValueError异常
        raise IterImproveError(guess)   # raise用于直接产生异常(是一个对象实例)，用于下面find_zero函数中执行except部分
```

```python
def newton_update(f,df):
    def update(x):
        x = x - f(x)/df(x)
        return x
    return update

def find_zero(f,df,guess=1):
    def done(x):
        return f(x) == 0
    try:
        return improve(newton_update(f,df),done,guess)
    except IterImproveError as e:   # 这个e绑定的是产生IterImproveError这个异常的对象实例
        return e.last_guess
```

```python
from math import sqrt
find_zero(lambda x: 2*x*x + sqrt(x),lambda x: 4*x + 0.5/sqrt(x))   
```
